# Agregación de datos por categoría

```python
import numpy as np
import pandas as pd
```

```python
gender= ['Male','Female']
income= ['Poor','Middle Class', 'Rich']
```

```python
n=500
gender_data=[]
income_data=[]

for i in range(0,n):
    gender_data.append(np.random.choice(gender))
    income_data.append(np.random.choice(income))
```

```python
gender_data[1:10]
```

```python
income_data[1:10]
```

```python
height= 160+30*np.random.randn(n)
weight= 65 +25*np.random.randn(n)
age= 20+12*np.random.randn(n)
income= 18000+3500*np.random.randn(n)
```

```python
data= pd.DataFrame(
    {
        'Gender': gender_data,
        'Economic Status': income_data,
        'Height': height,
        'weight': weight,
        'Age': age,
        'Income': income
        
    }
)
```

```python
data.head()
```

## Agrupacion de datos

No rompe o divide los datos, solo crea un objeto con dos atributos, el nombre y el grupo.

```python
grouped_gender=data.groupby("Gender")
```

```python
grouped_gender.groups
```

```python
for names, group in grouped_gender:
    print(names)
    print(group)

```

```python
grouped_gender.get_group('Female')
```

```python
double_group=data.groupby(['Gender','Economic Status'])
```

```python
len(double_group)
```

```python
for names, group in double_group:
    print(names)
    print(group)
```

## Operaciones sobre datos agrupados

```python
double_group.sum()
```

```python
double_group.mean()
```

```python
# Devuelve el tamaño
double_group.size()
```

```python
double_group.describe()
```

```python
group_income=double_group['Income']
group_income.describe()
```

```python
double_group.aggregate(
    {
        'Income': np.sum,
        'Age': np.mean,
        'Height': np.std
    }
)
```

```python
double_group.aggregate(
    {
        'Age': np.mean,
        'Height': lambda h:(np.mean(h))/np.std(h)
    }
)
```

```python
double_group.aggregate([np.sum,np.mean, np.std])
```

```python
double_group.aggregate([lambda x: (np.mean(x))/np.std(x)])
```

## Filtrado de datos

```python
double_group['Age'].filter(lambda x: x.sum()>1400)
```

## Tranformación de variables

```python
zscore =lambda x: (x-x.mean())/x.std()
```

```python
z_group=double_group.transform(zscore)
```

```python
import matplotlib.pyplot as plt
```

```python
plt.hist(z_group['Age'])
```

```python
fill_na_mean = lambda x: x.fillna(x.mean())
```

```python
double_group.transform(fill_na_mean)
```
